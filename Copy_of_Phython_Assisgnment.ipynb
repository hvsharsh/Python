```python
# Write a Python program to convert user input from string to integer, float, and boolean types.

#taking input from user
user_input = input("Enter a value: ")
#convert user input from string to integer, float, and boolean types.

try :
  integer_value = int(user_input)
except ValueError:
  integer_value = None
print("integer conversion ", integer_value)

try :
  float_value = float(user_input)
except ValueError:
  float_value = None
print("float conversion ", float_value)

boolean_value = bool(user_input)
print("boolean conversion ", boolean_value)

```

```python
#. Write code to demonstrate type casting with list elements.

list_str = ['10','20','30','40']

#type casting with list elements.
#convert to integers
list_int = [int(item) for item in list_str]
print("Integer list:", list_int)

#convert to float
list_float = [float(item) for item in list_str]
print("float list:", list_float)

#convert to boolean
list_num = [-1 , 2 , 0 , -6]
list_bool = [bool(item) for item in list_num]
print("boolean list:", list_bool)

```

```python
# Write a program to reverse a string using a while loop.

string = input("enter a string:")

#initialize variable
reversed_string = ""
i = len(string) - 1

#using while loop for reversing the string
while i >= 0:
  reversed_string += string[i]
  i -= 1
# display reversed sting
print("reversed string:", reversed_string)
```

```python
#Write a Python program to calculate the factorial of a number provided by the user using a while loop

num = int(input("enter a number"))
factorial = 1
i = 1
while i <= num :
  factorial *= i
  i += 1
  print("factorial of a " , num , "is:" , factorial)
```
