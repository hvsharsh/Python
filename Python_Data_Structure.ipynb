14.) How does the “in” keyword work differently for lists and dictionaries?
- "in" keyword work diffrently in list and dictionaries
    ##### LIST
    * For list, "in" keyword is used to check if the value is present in the list by scanning each elements in the list, It returns True if the value is present or gives False when value is not present.
    * ex:
    *     lis = [10, 20, 30]
    *     print(20 in lis)   # gives True
    *     print(80 in lis)   # gives False

    ##### Dictionary
    * When used with a dictionary, the "in" keyword checks for the presence of a specific key within the dictionary. It does not directly check for values
    * To check for the presence of a value in a dictionary, you need to explicitly use the .values()
    * To check for the presence of a key-value pair, you can use the .items() method
    * ex:
    *     person = {"name": "Alice", "age": 25}
    *     print("name" in person)  # True
    *     print("Alice" in person) # False

17.) Describe the time complexity of accessing elements in a dictionary
- Accessing elements in a Python dictionary has an average time complexity of O(1), meaning the retrieval time is constant regardless of the dictionary's size
this efficiency is because of hash table implimentation which uses the hash key to directly compute the memory location of the associated value.
#### Average Case (O(1))
- The hash table allows for direct access by mapping keys to specific locations, enabling constant time lookups.
#### Worst Case: O(n)
- In rare cases where many keys produce the same hash (hash collisions), Python handles these collisions using mechanisms like linked lists or balanced trees. This can degrade lookup time to linear in the number of keys. However, such cases are atypical in practical scenarios

18.)In what situations are lists preferred over dictionaries?
- list is preffered over dictionaries in several situations, primarily when order of element is important or when elements are accessed by their numerical position rather than by a unique key.
###### Here are specific scenario
* 1.) Ordered Collection -> when a sequence of elements matters, such as maintaining chronological order a queue (FIFO), or a stack (LIFO), lists are the appropriate choice. Dictionaries
* 2.) Access by Index -> If elements need to be retrieved, modified, or deleted based on their position within the collection, lists offer efficient O(1) time complexity for these operations. Dictionaries require a key for access.
* 3.) Storing collection of similar item -> When dealing with a collection of items that are conceptually similar and don't require unique identifiers for individual access, lists are a simple and effective way to store them. Examples include a list of numbers, names, or objects of the same type.
* 4.) Memory Efficiency -> Lists generally consume less memory than dictionaries because they only store values, whereas dictionaries store both keys and values, along with the overhead for hashing and managing key-value pairs.

19.) Why are dictionaries considered unordered, and how does that affect data retrieval?
- Dictionaries are considered as "unordered" because they are optimized for fast Key- Value based retrival not for maintaining specific sequence of elements and because of also their underlying implementation use hash table, which does not maintain any order of the key
- However in python(3.7+) dictionaries preserve insertion order as a language feature meaning the order in which keys are added is maintained when retrieving or iterating over items.
###### How it affect data retrival
* since dictionaries retrive data by key value pairs not fixed position or indexes, you cannot rely on ordering for algorithmic logic that depends on element sequence.
* Operations like sorting or accessing elements by numeric index are not applicable inherently to dictionaries (unlike lists), so you often need additional steps to sort keys or extract ordered data for specific use cases.
